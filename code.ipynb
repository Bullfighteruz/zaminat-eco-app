```
import os
import re

# Search for the HTML code snippet manually using basic file operations
def search_files_for_pattern(directory, pattern, file_extensions):
    """Search for a pattern in files with specific extensions"""
    found_files = []
    
    for root, dirs, files in os.walk(directory):
        for file in files:
            if any(file.endswith(ext) for ext in file_extensions):
                file_path = os.path.join(root, file)
                try:
                    with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
                        content = f.read()
                        if pattern in content:
                            found_files.append((file_path, content))
                except Exception as e:
                    continue
    
    return found_files

# Define search patterns and directory
src_directory = "/workspace/shadcn-ui/src"
file_extensions = ['.tsx', '.jsx', '.js', '.ts']

# Search for the overlay pattern
overlay_pattern = "absolute inset-0 flex items-center justify-center bg-black/20"
lock_pattern = "lucide-lock"

print("Searching for overlay pattern...")
overlay_files = search_files_for_pattern(src_directory, overlay_pattern, file_extensions)

print(f"Found {len(overlay_files)} files with overlay pattern:")
for file_path, content in overlay_files:
    print(f"\n--- File: {file_path} ---")
    # Extract relevant lines around the pattern
    lines = content.split('\n')
    for i, line in enumerate(lines):
        if overlay_pattern in line:
            # Show context around the match
            start = max(0, i-5)
            end = min(len(lines), i+10)
            print(f"Context around line {i+1}:")
            for j in range(start, end):
                marker = ">>> " if j == i else "    "
                print(f"{marker}{j+1:3}: {lines[j]}")
            print("-" * 50)

print("\n\nSearching for lock icon pattern...")
lock_files = search_files_for_pattern(src_directory, lock_pattern, file_extensions)

print(f"Found {len(lock_files)} files with lock icon:")
for file_path, content in lock_files:
    print(f"\n--- File: {file_path} ---")
    # Extract relevant lines around the pattern
    lines = content.split('\n')
    for i, line in enumerate(lines):
        if lock_pattern in line:
            # Show context around the match
            start = max(0, i-3)
            end = min(len(lines), i+8)
            print(f"Context around line {i+1}:")
            for j in range(start, end):
                marker = ">>> " if j == i else "    "
                print(f"{marker}{j+1:3}: {lines[j]}")
            print("-" * 30)
```
